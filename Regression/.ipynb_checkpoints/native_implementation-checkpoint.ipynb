```python
%matplotlib inline
import torch
from IPython import display
from matplotlib import pyplot as plt
import numpy as np
import random
```

# y=Xw+b+Ïµ 
#### linear regression equation

```python
num_inputs = 2 
num_examples = 1000 
true_w = [2, -3.4] 
true_b = 4.2 
features = torch.randn(num_examples, num_inputs, dtype=torch.float32) 
labels = true_w[0] * features[:, 0] + true_w[1] * features[:, 1] + true_b 
labels += torch.tensor(np.random.normal(0, 0.01, size=labels.size()), dtype=torch.float32) #add noise, epsilon
```

```python
def use_svg_display():
    #display by vector 
    display.set_matplotlib_formats('svg')
    
def set_figsize(figsize=(3.5,2.5)):
    use_svg_display()
    #set the dimensions of picture
    plt.rcParams['figure.figsize'] = figsize

set_figsize()
plt.scatter(features[:, 1].numpy(), labels.numpy(), 1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices) #samples should be read in a random order
    for i in range(0, num_examples, batch_size):
        j = torch.LongTensor(indices[i:min(i + batch_size, num_examples)]) #use min to check bound 
        #j is a indices vector of length batch_size
        yield features.index_select(0, j), labels.index_select(0, j) #return the 0th dimension index_select tensor
        
        
batch_size = 10

# for X,y in data_iter(batch_size, features, labels):
#     print('X y:',X,y)
#     break
    
# testTensor = torch.randn(3, 4, 4)
# print('test: ', testTensor)
# indexTensor = torch.tensor([0, 2])

# print(testTensor.index_select(2, indexTensor))

```

```python

w = torch.tensor(np.random.normal(0, 0.01, (num_inputs, 1)), dtype=torch.float32) #generate the random weight param
b = torch.zeros(1, dtype=torch.float32) #generate the random y intercept
 
w.requires_grad_(True)
b.requires_grad_(True)
```

#### On setting .requires_grad = True, the tensors start forming a backward graph that tracks every operation applied on them to calculate the gradients using something called a dynamic computation graph (DCG)

```python
def linreg(X, w, b):
    return torch.mm(X,w) + b

def squared_loss(yhat, y):
    #print(type(yhat), type(y))
    #print('yhat',yhat, 'y', y)
    return (yhat - y.view(yhat.size())) ** 2 / 2

def sgd(params, lr, batch_size): 
    for param in params:
        param.data -= lr * param.grad / batch_size 
```

# SGD Training

```python
lr = 0.03
num_epochs = 3
net = linreg  
loss = squared_loss #loss function

for epoch in range(num_epochs):
    #train the model with all the batches per epoch
    for X,y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y).sum()
        l.backward() 
        sgd([w, b], lr, batch_size) #update the parameters 
        
        #need to clear gradient values, since the default action for backward is to accumulate gradients
        w.grad.data.zero_()
        b.grad.data.zero_()
        
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f'%(epoch + 1, train_l.mean().item()))

print(true_w, '\n', w)
print(true_b, '\n', b)
        
```

### Backward is the function which actually calculates the gradient by passing itâs argument (1x1 unit tensor by default) through the backward graph all the way up to every leaf node traceable from the calling root tensor. 

### The method is to calculate the derivative and plug in the number as the original tensor's value, and write the value to the grad.data to the leaf nodes (tensors) and to the internal nodes if retain_grad is True. 

### the grad_fn of the tensor marks the last function executed on the tensor
